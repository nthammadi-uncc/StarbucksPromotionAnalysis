```python
#Importing packages
import pandas as pd
import numpy as np
```

```python
#Loading Data and viewing it
df = pd.read_csv('/final_consolidated_offers.csv')
df
```

```python
df = df.drop(['gender','became_member_on','offer_type','c_record_counts', 't_record_counts', 'v_record_counts'], axis=1)
df
```

```python
df.dtypes(   
```

```python
df.isna().sum()
```

```python
df = df.fillna(0)
```

```python
df = df.replace('UNKNOWN', np.nan, regex=True)
df = df.replace('', np.nan, regex=True)
```

```python
#Checking for nulls
df.isna().sum()
```

```python
#Imputing NA/s with mean
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(fill_value=np.nan, strategy='mean')
df1 = imputer.fit_transform(df)
df1 = pd.DataFrame(df1, columns = df.columns)
df1
```

```python
df1.isna().sum()
```

```python
#Standardizing data using StandardScaler
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
y = df1['successful_offer']
y =y.astype('int')
X = df1.drop(['successful_offer'], axis=1)

X = pd.DataFrame(ss.fit_transform(X), columns= X.columns)
```

```python
#Splitting our model into train and test sets using Train Test Split to make 20% of our data testing and 80% training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
import sklearn.linear_model
logreg = sklearn.linear_model.LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
predictions = logreg.predict(X_test)
score = logreg.score(X_test, y_test)
print (score)
print(logreg.coef_)
```
