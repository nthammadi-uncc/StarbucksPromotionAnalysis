## Exploratory Data Analysis

### Load necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

### Read data files

```python
#read the csv files and convert into dataframe
parent_url='https://raw.githubusercontent.com/nthammadi-uncc/StarbucksPromotionAnalysis/main/data/clean/'
offer_received_url=parent_url+'offer_received.csv'
offer_viewed_url=parent_url+'offer_viewed.csv'
offer_completed_url=parent_url+'offer_completed.csv'

offer_received_df=pd.read_csv(offer_received_url)
offer_received_df = offer_received_df.iloc[: , 1:]
offer_viewed_df=pd.read_csv(offer_viewed_url)
offer_viewed_df = offer_viewed_df.iloc[: , 1:]
offer_completed_df=pd.read_csv(offer_completed_url)
offer_completed_df = offer_completed_df.iloc[: , 1:]
```
